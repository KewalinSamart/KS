```python
given_data_x = [1.0,1.1,1.3,1.5,1.9,2.1]
```

```python
given_data_y = [1.84,1.96,2.21,2.45,2.94,3.18]
```

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0.75,2.25,num=100)
P1 = []
P2 = []
P3 = []
for i in range(len(x)):
    P1.append(0.620895+1.29621*x[i])
    P2.append(0.596581+1.25329*x[i]-0.0108534*x[i]**2)
    P3.append(0.629019+1.18501*x[i]+0.0353325*x[i]**2-0.0100472*x[i]**3)
```

```python
xx = np.linspace(0.75,20,num=100)
P1x = []
P2x = []
P3x = []
for i in range(len(xx)):
    P1x.append(0.620895+1.29621*xx[i])
    P2x.append(0.596581+1.25329*xx[i]-0.0108534*xx[i]**2)
    P3x.append(0.629019+1.18501*xx[i]+0.0353325*xx[i]**2-0.0100472*xx[i]**3)
```

```python
plt.plot(x,P1,label="P1(x)")
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 1 (zoom in)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

plt.plot(xx,P1x,label="P1(x)")
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 1 (zoom out)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,P2,label="P2(x)",color='red')
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 2 (zoom in)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

plt.plot(xx,P2x,label="P2(x)",color='red')
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 2 (zoom out)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

```python
plt.plot(x,P3,label="P3(x)",color='black')
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 3 (zoom in)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()

plt.plot(xx,P3x,label="P3(x)",color='black')
plt.scatter(given_data_x,given_data_y,color='orange',label="actual data")
plt.title("The least squares polynomial degree 3 (zoom out)")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```
