### Write a Python code for Steepest Descent Method to solve 10.4 (1.b, 2.d).

```python
import numpy as np
```

#### Exercise set 10.4 (1.b)

```python
# define the given nonlinear system
def f1(x1,x2):
    f1 = 3*x1**2 - x2**2
    return f1

def f2(x1,x2):
    f2 = 3*x1*x2**2 - x1**3 -1
    return f2
#-----------------------------------#
def df1_x1(x1,x2): # f1' respect to x1
    df1_x1 = 6*x1
    return df1_x1

def df1_x2(x1,x2): # f1' respect to x2
    df1_x2 = -2*x2
    return df1_x2
#-----------------------------------#
def df2_x1(x1,x2): # f2' respect to x1
    df2_x1 = 6*x1
    return df2_x1

def df2_x2(x1,x2): # f2' respect to x2
    df2_x2 = 3*x1*2*x2
    return df2_x2
```

```python
def x_2D(x1,x2): # construct 2D vector <x1,x2>
    x = np.array([x1,x2])
    return x

def g_2D(x):
    x1 = x[0]
    x2 = x[1]
    g = (f1(x1,x2))**2 + (f2(x1,x2))**2 
    return g

# output: z
def gradient2D(x):
    x1 = x[0]
    x2 = x[1]
    z = 2 * np.array([f1(x1,x2)*df1_x1(x1,x2) + f2(x1,x2)*df2_x1(x1,x2), 
                               f1(x1,x2)*df1_x2(x1,x2) + f2(x1,x2)*df2_x2(x1,x2)])
    return z 
```

```python
# take gradient_g (z) 
# output: z0
def two_norm(z):
    result = 0
    for element in z:
        result += element**2
    z0 = np.sqrt(result)
    return z0
```

```python
# Steepest Descent (2D)
# Algorithm 10.3
def steepest_descent2D(x,N,TOL):
    k = 1
    while k <= N:
    
        g1 = g_2D(x)
        z = gradient2D(x)
        z0 = two_norm(z)
        if z0 == 0:
            print('Zero gradient')
            return([x,g1]) 
        
        z = z/z0 # make z a unit vector
        alpha1 = 0
        alpha3 = 1
        g3 = g_2D(x-alpha3*z)
    
        while g3 >= g1:
            alpha3 = alpha3/2
            g3 = g_2D(x-alpha3*z)
            
            if alpha3 < TOL/2:
                print("No likely improvement")
                return([x,g1])
            
        alpha2 = alpha3/2
        g2 = g_2D(x-alpha2*z)
        h1 = (g2 - g1)/alpha2
        h2 = (g3 - g2)/(alpha3-alpha2)
        h3 = (h2-h1)/alpha3
        
        alpha0 = 0.5*(alpha2-h1/h3) 
        g0 = g_2D(x-alpha0*z)
    
        # optimization of alpha to minimize g
        g0 = g_2D(x-alpha0*z)
        g3 = g_2D(x-alpha3*z)  
            
        if g0 < g3:
            g = g0
            opt_alpha = alpha0
        elif g0 > g3:
            g = g3
            opt_alpha = alpha3
                
        x = x-opt_alpha*z
        if abs(g-g1) < TOL:
            return([x,g1])        
            
        k += 1
    return [x,g1]
```

```python
TOL = 0.05
x1 = 1
x2 = 1
x = x_2D(x1,x2)
N = 30

r = steepest_descent2D(x,N,TOL)
print("Approximate x <x1, x2>:", r[0], "g:", r[1])
```

#### Exercise set 10.4 (2.d)

```python
# define the given nonlinear system
def h1(x1,x2,x3):
    h1 = x1 + np.cos(x1*x2*x3) - 1
    return h1

def h2(x1,x2,x3):
    h2 = (1-x1)**(1/4) + x2 + 0.05*x3**2 - 0.15*x3 - 1
    return h2

def h3(x1,x2,x3):
    h3 = -x1**2 - 0.1*x2**2 + 0.01*x2 + x3 -1
    return h3
#----------------------------------------------------------#
def dh1_x1(x1,x2,x3): # h1' respect to x1
    dh1_x1 = 1 - x2*x3*np.sin(x1*x2*x3)
    return dh1_x1

def dh1_x2(x1,x2,x3): # h1' respect to x2
    dh1_x2 = - x1*x3*np.sin(x1*x2*x3)
    return dh1_x2

def dh1_x3(x1,x2,x3): # h1' respect to x3
    dh1_x3 = - x1*x2*np.sin(x1*x2*x3)
    return dh1_x3
#----------------------------------------------------------#
def dh2_x1(x1,x2,x3): # h2' respect to x1
    dh2_x1 = -1/4*(1-x1) 
    return dh2_x1

def dh2_x2(x1,x2,x3): # h2' respect to x2
    dh2_x2 = 0
    return dh2_x2

def dh2_x3(x1,x2,x3): # h2' respect to x3
    dh2_x3 = 0.1*x3 - 0.15
    return dh2_x3
#----------------------------------------------------------#
def dh3_x1(x1,x2,x3): # h3' respect to x1
    dh3_x1 = -2*x1
    return dh3_x1

def dh3_x2(x1,x2,x3): # h3' respect to x2
    dh3_x2 = -0.2*x2 + 0.01
    return dh3_x2

def dh3_x3(x1,x2,x3): # h3' respect to x3
    dh3_x3 = 1
    return dh3_x3
```

```python
def x_3D(x1,x2,x3): # construct 3D vector  <x1,x2,x3>
    x = np.array([x1,x2,x3])
    return x

def g_3D(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    g = (h1(x1,x2,x3))**2 + (h2(x1,x2,x3))**2 + (h3(x1,x2,x3))**2 
    return g

# output: z
def gradient3D(x):
    x1 = x[0]
    x2 = x[1]
    x3 = x[2]
    z = 2 * np.array([h1(x1,x2,x3)*dh1_x1(x1,x2,x3) + h2(x1,x2,x3)*dh2_x1(x1,x2,x3) + h3(x1,x2,x3)*dh3_x1(x1,x2,x3), 
                    h1(x1,x2,x3)*dh1_x2(x1,x2,x3) + h2(x1,x2,x3)*dh2_x2(x1,x2,x3) + h3(x1,x2,x3)*dh3_x2(x1,x2,x3),
                     h1(x1,x2,x3)*dh1_x3(x1,x2,x3) + h2(x1,x2,x3)*dh2_x3(x1,x2,x3) + h3(x1,x2,x3)*dh3_x3(x1,x2,x3)])
    return z 
```

```python
# Steepest Descent (3D)
# Algorithm 10.3
def steepest_descent3D(x,N,TOL):
    k = 1
    while k <= N:
    
        g1 = g_3D(x)
        z = gradient3D(x)
        z0 = two_norm(z)
        if z0 == 0:
            print('Zero gradient')
            return([x,g1]) 
        
        z = z/z0 # make z a unit vector
        alpha1 = 0
        alpha3 = 1
        g3 = g_3D(x-alpha3*z)
    
        while g3 >= g1:
            alpha3 = alpha3/2
            g3 = g_3D(x-alpha3*z)
            
            if alpha3 < TOL/2:
                print("No likely improvement")
                return([x,g1])
            
        alpha2 = alpha3/2
        g2 = g_2D(x-alpha2*z)
        h1 = (g2 - g1)/alpha2
        h2 = (g3 - g2)/(alpha3-alpha2)
        h3 = (h2-h1)/alpha3
        
        alpha0 = 0.5*(alpha2-h1/h3) 
        g0 = g_3D(x-alpha0*z)
    
        # optimization of alpha to minimize g
        g0 = g_3D(x-alpha0*z)
        g3 = g_3D(x-alpha3*z)  
            
        if g0 < g3:
            g = g0
            opt_alpha = alpha0
        elif g0 > g3:
            g = g3
            opt_alpha = alpha3
                
        x = x-opt_alpha*z
        if abs(g-g1) < TOL:
            return([x,g1])        
            
        k += 1
    return [x,g1]
```

```python
TOL = 0.05
x1 = 0
x2 = 0
x3 = 0
x = x_3D(x1,x2,x3)
N = 30

r = steepest_descent3D(x,N,TOL)
print("Approximate x <x1, x2, x3>:", r[0], "g:", r[1])
```
