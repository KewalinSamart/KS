```python
import numpy as np
import matplotlib.pyplot as plt
```

### Legendre polynomials

```python
xp = np.linspace(-1,1,num=100)
P0, P1, P2, P3, P4 = ([] for i in range(5)) 
for i in range(len(xp)):
    P0.append(1)
    P1.append(xp[i])
    P2.append(1/2*(3*xp[i]**2-1))
    P3.append(1/2*(5*xp[i]**3-3*xp[i]))
    P4.append(1/8*(35*xp[i]**4-30*xp[i]**2+15*xp[i]))
```

```python
plt.plot(xp,P0,label="P0")
plt.plot(xp,P1,label="P1")
plt.plot(xp,P2,label="P2")
plt.plot(xp,P3,label="P3")
plt.plot(xp,P4,label="P4")

plt.title("Legendre polynomials")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

### Chebyshev polynomials

```python
xt = np.linspace(-1,1,num=100)
T0, T1, T2, T3, T4 = ([] for i in range(5)) 
for i in range(len(xt)):
    T0.append(1)
    T1.append(xt[i])
    T2.append(2*xt[i]**2-1)
    T3.append(4*xt[i]**3-3*xt[i])
    T4.append(8*xt[i]**4-8*xt[i]**2+1)
```

```python
plt.plot(xt,T0,label="T0")
plt.plot(xt,T1,label="T1")
plt.plot(xt,T2,label="T2")
plt.plot(xt,T3,label="T3")
plt.plot(xt,T4,label="T4")

plt.title("Chebyshev polynomials")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid()
plt.show()
```

### Comparing the properties
*Note: Denote n as the degree of the polynomials and $w(x)$ as the weight function.*
* **Legendre:** 
    - $x \in [-1,1]$
    - $w(x) = 1$
    - As n increases, the global maximum goes toward $\infty$ and the global minimum goes towards $-\infty$.
    - The function always crosses zero on the x axis n times.
    - its global maximum $\neq$ its local maximum for n > 0.
    - its global miniimum $\neq$ its local miniimum for n > 0.
    
* **Chebychev:**
    - $x \in [-1,1]$
    - $w(x) = \frac{1}{\sqrt{1-x^2}}$
    - no matter what n is, the function always oscollates between -1 and 1.
    - The function always crosses zero on the x axis n times.
    - its global maximum = its local maximum = 1 for n > 0.
    - its global minimum = its local minimum = -1 for n > 0.
