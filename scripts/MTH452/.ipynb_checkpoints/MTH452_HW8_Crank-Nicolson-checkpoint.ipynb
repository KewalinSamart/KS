```python
import numpy as np
```

```python
# 12.2 (1)
m = 4
T = 0.1
N = 2
l = 2
alpha_sq = 1
```

```python
# 12.2 (1)
def f(x):
    f_x = np.sin(np.pi/2*x)
    return f_x

def exact_sol(x,t):
    sol = np.exp(-((np.pi**2)/4)*t)*np.sin(np.pi/2*x)
    return sol
```

```python
# 12.2 (2)
m = 3
T = 0.1
N = 2
l = 1
alpha_sq = 1/16
```

```python
# 12.2 (2)
def f(x):
    f_x = 2*np.sin(2*np.pi*x)
    return f_x

def exact_sol(x,t):
    sol = 2*np.exp(-((np.pi**2)/4)*t)*np.sin(2*np.pi*x)
    return sol
```

```python
# Algorithm 12.3: Crank-Nicolson
def Crank_Nicolson(m, T, N, l, alpha_sq):
    h = l/m
    k = T/N
    ld = alpha_sq*k/h**2
    wm = 0
    wi = [None] # set index 0 to None for convenience in indexing

    for i in range(1,m):
        wi.append(f(i*h))
    wi.append(wm)   
    print()
    li = [None]
    ui = [None]
    li.append(1 + ld)
    ui.append(-1*ld/(2*li[1]))

    for i in range(2,m-1):
        li.append(1 + ld + ld*ui[i-1]/2)
        ui.append(-ld*(2*li[i]))

    li.append(1 + ld + ld*ui[m-2]/2)

    zi = [None]
    t_ = []
    x_ = []
    print('i, x, t, wi, exact sol')
    for j in range(1,N+1):
        t = j*k    # (t at the index j)
        t_.append(t)
        zi.append(((1-ld)*wi[1] + ld/2 * wi[2])/li[1])
            
        for i in range(2,m):
            zi.append(((1-ld)*wi[i] + ld/2*(wi[i+1] + wi[i-1] + zi[i-1]))/li[i])
        wi[m-1] = zi[m-1]
    
        for i in reversed(range(1,m-1)):
            wi[i] = zi[i] - ui[i]*wi[i+1]
            xi = i*h
            if len(x_) < N:
                x_.append(x)
            print(i,xi,t,wi[i],exact_sol(xi,t))
```

```python
# 12.2 (1)
Crank_Nicolson(m, T, N, l, alpha_sq)
```

```python
# 12.2 (2)
Crank_Nicolson(m, T, N, l, alpha_sq)
```
