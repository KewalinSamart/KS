```python
import numpy as np
```

```python
m = 4
N = 4
T = 0.5
l = 0.5
alpha = 1/(4*np.pi)
```

```python
def f(x):
    return 0

def g(x):
    g_ = np.sin(4*np.pi*x)
    return g_

def u(x,t):
    u_ = np.sin(t)*np.sin(4*np.pi*x)
    return u_
```

```python
# Wave Equation Finite-Difference
# Algorithm 12.4
h = l/m
k = T/N
ld = k*alpha/h
w = [[None]*(m+1)]*(N+1)
for j in range(1,N+1):
    w[0][j] = 0
    w[m][j] = 0
w[0][0] = f(0)
w[m][0] = f(l)
for i in range(1,m):
    w[i][0] = f(i*h)
    w[i][1] = (1-ld**2)*f(i*h) + (ld**2)/2*(f((i+1)*h) + f((i-1)*h)) + k*g(i*h)
for j in range(1,N):
    for i in range(1,m):
        w[i][j+1] = 2*(1-ld**2)*w[i][j] + (ld**2)*(w[i+1][j] + w[i-1][j]) - w[i][j-1]
print('x, t, wij, u(xi,tj)')
for j in range(0,N+1):
    t = j*k
    if t == 0.5:
        for i in range(0,m+1):
            x = i*h
            print(x, t, w[i][j], u(x,t), abs(w[i][j]-u(x,t)))
    
```
